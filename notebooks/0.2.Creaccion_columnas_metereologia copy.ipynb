## Importar librerías


```python
import pandas as pd
import numpy as np
import os
from data_util import calc_missing
```

## Cargar datos


```python
file_names = os.listdir('../data/metereologia/')
file_names
```

## Crear un único dataset con los datos del año 2022

```python
df_meteo = pd.DataFrame()
for file_name in file_names:
    df_current = pd.read_csv('../data/metereologia/' + file_name, delimiter=';')
    print(file_name, df_current.shape)
    df_meteo = pd.concat([df_meteo, df_current], ignore_index=True)
df_meteo.head(3)
```

## Nos quedamos con los parámetros de las estaciones que nos interesan:

8, 18, 24, 35, 36

```python
# df_meteo  = df_meteo[(df_meteo['ESTACION'] == 8) | (df_meteo['ESTACION'] == 18) | (df_meteo['ESTACION'] == 24 ) |
#                     (df_meteo['ESTACION'] == 35) | (df_meteo['ESTACION'] == 36)]
df_meteo  = df_meteo[(df_meteo['ESTACION'] == 8)]
df_meteo.head()
```

# Nos quedamos con los parámetros que nos interesan: 
 velocidad del viento, la temperatura, la humedad relativa y las precipitaciones.


```python
df_meteo = df_meteo[(df_meteo['MAGNITUD'] == 81) | (df_meteo['MAGNITUD'] == 82) | (df_meteo['MAGNITUD'] == 83) | (df_meteo['MAGNITUD'] == 86) | (df_meteo['MAGNITUD'] == 89)]
df_meteo.head()
```

```python
df_meteo.info()
```

## Preparación de los datos de salida

```python
data_columns = [ 'H01', 'H02', 'H03', 'H04', 'H05', 'H06', 'H07', 'H08', 'H09', 'H10', 'H11', 'H12', 'H13', 'H14', 'H15', 'H16', 'H17', 'H18', 'H19', 'H20', 'H21', 'H22', 'H23', 'H24'] # feature
veri_columns = [ 'V01', 'V02', 'V03', 'V04', 'V05', 'V06', 'V07', 'V08', 'V09', 'V10', 'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20', 'V21', 'V22', 'V23', 'V24']

hora_column = np.arange(1,25) # feature label

data = pd.DataFrame(columns=['HORA', 'DATO', 'VERIFICADO'])
aux = pd.DataFrame(columns=['ANO', 'MES', 'DIA', 'HORA', 'DATO', 'VERIFICADO', 'ESTACION']) # aux de próxima fila a ser añadida
for  (h,d,v) in zip(hora_column, data_columns, veri_columns): 
    aux
    aux['DATO'] = df_meteo[ data_columns[h-1] ] 
    aux[['CODIGO', 'MAGNITUD', 'TECNICA']] = df_meteo['PUNTO_MUESTREO'].str.split('_', n=2, expand=True)
    aux['VERIFICADO'] = df_meteo[ veri_columns[h-1] ] 
    aux['HORA'] = h-1
    aux[['ANO', 'MES', 'DIA']] = df_meteo[['ANO','MES','DIA']]
    #aux['MAGNITUD'] = df_aire ['MAGNITUD']
    aux['ESTACION'] = df_meteo ['ESTACION']

    data = pd.concat([data, aux], ignore_index=True)
data = data.astype({'HORA': 'int32'}, copy = False)

data.head()
```

# Añadimos columna fecha de formato datetime

```python
data['FECHA'] = pd.to_datetime(data[['ANO','MES','DIA','HORA']].rename(columns = {'ANO': 'year', 'MES': 'month','DIA': 'day', 'HORA': 'hour'}))
data.head()
```

# Ordenamos datos por orden cronológico

```python
data = data.sort_values(['FECHA','HORA'])
data.head()
```

## Marcar datos no válidos en el DataFrame

Datos no verificados

```python
data.loc[data['VERIFICADO'] != 'V', ['DATO']] = np.nan
data.loc[data['VERIFICADO'] != 'V']
```

# Borramos columna verificado

```python
data = data.drop('VERIFICADO', axis=1)
data.info()
```

## Añadir columnas que faltan y formatearlas

```python
#data = data.set_index(['CODIGO','FECHA', 'ANO', 'MES', 'DIA', 'HORA', 'ESTACION','MAGNITUD'])['DATO'].unstack(level=-1).reset_index()
data = data.set_index(['FECHA','MAGNITUD'])['DATO'].unstack(level=-1).reset_index()

data.head()
```
